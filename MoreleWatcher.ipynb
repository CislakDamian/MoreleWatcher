```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import os
from selenium.webdriver.firefox.options import Options as FirefoxOptions

options = FirefoxOptions()
options.add_argument("--headless")

driver = webdriver.Firefox(options=options)
driver.get('https://www.morele.net/')

search_box = driver.find_element_by_name('search')
search_box.send_keys('redmi note 9')
search_box.send_keys(Keys.ENTER)
time.sleep(5)

list_of_item = driver.find_elements_by_class_name('productLink')
name = [title.text for title in list_of_item]
while("" in name) : 
     name.remove("") 

price_class = driver.find_elements_by_class_name('price-new')
price  = [title.text for title in price_class]
while("" in price) : 
    price.remove("") 

driver.close()

price = pd.DataFrame(price) 
name = pd.DataFrame(name) 
name.columns = ['desc']
price.columns = ['price']

df = pd.concat([name, price],  axis=1,sort=False)
df = df[df.desc.str.contains("Redmi", na=False)]
df['pull_datetime'] = pd.Timestamp.now().strftime("%Y-%m-%d %H:%M:%S")
df['currency'] = 'PLN'
df['price'] = df['price'].str.replace(' z≈Ç', '')
df = df[['pull_datetime','desc','price','currency']]
df['found_flag'] = ''
df['found_flag'][df['desc'].str.contains('Note 9')] = '1'

hdr = False  if os.path.isfile('Morele.csv') else True
df.to_csv('Morele.csv', mode='a', header=hdr, index=False)
```
